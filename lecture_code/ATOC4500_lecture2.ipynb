```python
## Example #1

import scipy.stats as stats
import numpy as np

print('% data in between -1 and 0 for a standard normal distribution')
value=stats.norm.cdf(1,loc=0, scale=1)-stats.norm.cdf(0,loc=0, scale=1)
print(value*100)
print(np.round(value*100))
print('% data NOT in between -1 and 0 for a standard normal distribution')
print(np.round((1-value)*100))
```

```python
## Example #2

import numpy as np
import matplotlib.pyplot as plt

# draw random values from normal distribution
x = np.random.normal(0,1,size = (5,))   ## small sample (N=5)
y = np.random.normal(0,1,size = (3000,)) ## big sample (N=3000)

# calculate histogram
bins = np.arange(-4.5,5.5,.5)
hx = np.histogram(x,bins)
hy = np.histogram(y,bins)

## plot Normalized Probability Distribution Function
plt.xlabel('Value')
plt.ylabel('Frequency (%)')
plt.bar(hy[1][:-1],(hy[0].astype(float)/np.size(y))*100,edgecolor='r', color=[], 
        width = .4, linewidth = 2, label='y');
plt.bar(hx[1][:-1],(hx[0].astype(float)/np.size(x))*100,edgecolor='k', color=[], 
        width = .4, linewidth = 2, label='x');
plt.show() ## needed if not the last plot in a cell
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Plot Cumulative Distribution Function
hxx = np.cumsum(hx[0])
hyy = np.cumsum(hy[0])
plt.xlabel('Value')
plt.ylabel('Frequen(cy (%)')
plt.bar(hx[1][:-1],(hxx.astype(float)/np.size(x))*100,edgecolor='k', 
        color=[], width = .4, linewidth = 2);
plt.bar(hy[1][:-1],(hyy.astype(float)/np.size(y))*100,edgecolor='r', 
        color=[], width = .4, linewidth = 2);
```

```python
## Example #3

import numpy as np
N=5
## print five random values drawn from a normal distribution
values=np.random.normal(loc = 0, scale = 1., size = N)
print(values)
## calculate and print the mean of these random values
print('mean=',np.round(np.mean(values)),2)
## calculate and print the standard deviation of these random values
print('std. dev.=',np.round(np.std(values),2))
```
